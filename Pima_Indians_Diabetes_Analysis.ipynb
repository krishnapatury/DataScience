# Diabetes Data Analysis


Diabetes is one of the most frequent diseases worldwide and the number of diabetic patients are growing over the years. The main cause of diabetes remains unknown, yet scientists believe that both genetic factors and environmental lifestyle play a major role in diabetes.

A few years ago research was done on a tribe in America which is called the Pima tribe (also known as the Pima Indians). In this tribe, it was found that the ladies are prone to diabetes very early. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients were females at least 21 years old of Pima Indian heritage. 


## Objective
-----------------

Exploratory Data Analysis to understand different aspects of Diabetes in the Pima Indians tribe.

## Data Dictionary
-------------------------

The dataset has the following information:

* Pregnancies: Number of times pregnant
* Glucose: Plasma glucose concentration over 2 hours in an oral glucose tolerance test
* BloodPressure: Diastolic blood pressure (mm Hg)
* SkinThickness: Triceps skin fold thickness (mm)
* Insulin: 2-Hour serum insulin (mu U/ml)
* BMI: Body mass index (weight in kg/(height in m)^2)
* DiabetesPedigreeFunction: A function that scores the likelihood of diabetes based on family history.
* Age: Age in years
* Outcome: Class variable (0: a person is not diabetic or 1: a person is diabetic)

## Mean, median and mode of the variable 'BMI'. Observations.

```python
m1 = pima['BMI'].mean()  # mean
print("The mean of the variable 'BMI' is ", m1)

m2 = pima['BMI'].median()  # median
print("The median of the variable 'BMI' is ", m2)

m3 = pima['BMI'].mode()[0]  # mode
print("The mode of the variable 'BMI' is ", m3)
```

The mean, median, and mode are all measures of central tendency. While the median and mode are exactly equal, the mean is slightly higher than them. So even though BMI is a normal distribution it will be slightly right skewed.

## Interquartile Range of all the variables.

```python
Q1 = pima.quantile(0.25)
Q3 = pima.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

The interquartile range is the middle half of the data that lies between the upper and lower quartiles, i.e, it includes the 50% of data points that are above Q1 and below Q4. When measuring variability, statisticians prefer using the interquartile range instead of the full data range because extreme values and outliers affect it less. Values which are 1.5*IQR less than Q1 or 1.5*IQR more than Q3 may be considered outliers and ignored in caclucaltions based on the problem. Typically, we use the IQR with a measure of central tendency, such as the median, to understand the datas center and spread.

## The correlation matrix

```python
corr_matrix = pima.iloc[ : ,0 : 8].corr()

plt.figure(figsize = (8, 8))
sns.heatmap(corr_matrix, annot = True)

# Display the plot
plt.show()
```

Pregnancies and Age and Skin Thickness and BMI seem to be positively correlated with correlation coefficients of 0.54 and 0.53 respectively indicating that the number of pregnancies are higher in the older women and that BMI is higher for those with higher skin thickness. While Glucose and Age, Blood pressure and Age, Glucose and BMI, Blood Pressure and BMI, and Skin Thickness and insulin seem to have a slight positive correlation their values are too small to indicate a correlation. Glucose and Insulin also seem to some positive correlation. The remaining atrributes **do** not have a very strong correlation with each other either positive or negative.
